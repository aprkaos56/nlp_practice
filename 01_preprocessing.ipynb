```python
# 뉴스 기사 크롤링
import requests
from bs4 import BeautifulSoup

url = "https://www.newsis.com/view/NISX20260226_0003526930" # "드디어 오르나"…비트코인, 1억 돌파 눈앞
response = requests.get(url)

bs = BeautifulSoup(response.text, 'html.parser')

article = bs.select_one('article')
text = article.get_text('\n', strip=True)

print(text)
```

```python
# 문장 토큰화
import kss

kss.split_sentences(text)
```

```python
# 불용어 제거
from konlpy.tag import Okt

okt = Okt()
tokens = okt.morphs(text)
print(len(tokens))
print(tokens)
```

```python
def load_ko_stopwords(filepath):
    with open(filepath, 'r', encoding='UTF-8') as f:
        return [line.strip() for line in f]
    
ko_stopwords = load_ko_stopwords('ko_stopwords.txt')
cleaned_tokens = [token for token in tokens if token not in ko_stopwords]
print(len(cleaned_tokens))
print(cleaned_tokens)
```

```python
result = " ".join(cleaned_tokens)
result
```

```python
# 평가
from nltk.sentiment.vader import SentimentIntensityAnalyzer

analyzer = SentimentIntensityAnalyzer()
analyzer.polarity_scores(result)
```
