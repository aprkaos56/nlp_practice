```python
# 데이터 불러오기
import pandas as pd

df = pd.read_csv('DisneylandReviews.csv')
df_HK = df[df['Branch'] == 'Disneyland_HongKong']    # Disneyland_HongKong만 가져오기

texts = df_HK['Review_Text'].astype(str).str.lower().tolist()
text = " ".join(texts)
len(text)
```

```python
from nltk.tokenize import RegexpTokenizer

# 알파벳만 추출
tokenizer = RegexpTokenizer(r"[a-zA-Z]+")

# 토큰화
tokens = tokenizer.tokenize(text)
len(tokens)
```

```python
# 짧은 단어 제거
tokens = [token for token in tokens if len(token) > 2]
len(tokens)
```

```python
from collections import Counter

# 빈도 수 낮은 단어 제거
word_counts = Counter(tokens)

tokens = [token for token in tokens if word_counts[token] >= 2]
len(tokens)
```

```python
from nltk.corpus import stopwords

# 불용어 제거
en_stopwords = stopwords.words('english')

tokens = [token for token in tokens if token not in en_stopwords]
len(tokens)
```

```python
result = " ".join(tokens)
result[:500]
```

```python
# 평가
from nltk.sentiment.vader import SentimentIntensityAnalyzer

analyzer = SentimentIntensityAnalyzer()
analyzer.polarity_scores(result)
```
